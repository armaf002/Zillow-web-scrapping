```python
from bs4 import BeautifulSoup                          
import requests                                        
import re
import time
import pandas as pd

def dataframe(zillow_property_url):
    req_headers = {
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'accept-encoding': 'gzip, deflate, br',
    'accept-language': 'en-US,en;q=0.8',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'
    }         
    res = requests.get(url=zillow_property_url, headers=req_headers)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    title = soup.title.text                             
    address = title.split('|')[0].strip()   
    site = title.split('|')[2].strip()
    
    zillow_property_ID = zillow_property_url.split('/')[-2].strip()
    zillow_property_ID = zillow_property_ID.split('_zpid')[0].strip()
    zillow_property_ID = int(zillow_property_ID)

    buyer_agency_compensation = soup.findAll('span', {'class': 'Text-c11n-8-62-5__sc-aiai24-0 kZKvMY'})
    buyer_agency_compensation = buyer_agency_compensation
    buyer_agency_compensation = ' '.join(map(str, buyer_agency_compensation))
    buyer_agency_compensation = buyer_agency_compensation.split('Buyer agency compensation<!-- -->: <!-- -->')[1].strip()
    buyer_agency_compensation = buyer_agency_compensation.split('</span>')[0].strip()
    buyer_agency_compensation = re.sub('[%]','', buyer_agency_compensation)
    buyer_agency_compensation = float(buyer_agency_compensation)/100

    annual_tax_amount = soup.findAll('span', {'class': 'Text-c11n-8-62-5__sc-aiai24-0 kZKvMY'})
    annual_tax_amount = ' '.join(map(str, annual_tax_amount))
    annual_tax_amount = annual_tax_amount.split('Annual tax amount<!-- -->: <!-- -->$')[1].strip()
    annual_tax_amount = annual_tax_amount.split('</span>')[0].strip()
    annual_tax_amount = re.sub('[,]','', annual_tax_amount)
    annual_tax_amount = float(annual_tax_amount)

    
    buyer_agent_fee = soup.findAll('span', {'class':'Text-c11n-8-62-5__sc-aiai24-0 dpf__sc-2arhs5-3 kZKvMY btxEYg'})
    buyer_agent_fee = ' '.join(map(str, buyer_agent_fee))
    buyer_agent_fee = buyer_agent_fee.split('price/sqft</span> <span class="Text-c11n-8-62-5__sc-aiai24-0 dpf__sc-2arhs5-3 kZKvMY btxEYg">')[1].strip()
    buyer_agent_fee = buyer_agent_fee.split('%<!-- -->')[0].strip()
    buyer_agent_fee = float(buyer_agent_fee)/100
    
    information = soup.find('span', {'class': 'Text-c11n-8-62-5__sc-aiai24-0 hdp__sc-b5iact-0 frfoXM fAzOKk'}).text
    price = information.split('$')[1].strip()
    price = re.sub('[,]','',  price)
    price = float(price)

    info = {'zillow_proper_ID': [zillow_property_ID],
    'zillow_property_url': [zillow_property_url],
       'buyerAgentFee': [buyer_agent_fee], 
       'buyerAgencyCompensation': [buyer_agency_compensation],
       'annualTaxAmount': [annual_tax_amount],
       'price': [price],
       'address': [address],
       'site': [site]}

    dataframe = pd.DataFrame(info)
    return dataframe    




    
    



df = dataframe('https://www.zillow.com/homedetails/3320-Galloway-Dr-San-Diego-CA-92122/16844858_zpid/')
df
```
